#Dropping irrelevant columns

```python
col='Postal Code'
data1=data.drop(columns=col,axis=1)
data1.head()
```

#STEP 4 - Checking the statistical relationship between various rows and columns

```python
#Correlation between variables
correlation=data1.corr()
correlation
```

#Exploratory Data Analysis

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

From the above heatmap we can see that 
1.Profit and sales are positively correlated.(As sales increases profit increases)
2.Quantity and profit are positively correlated.(As quantity increases profit increases)
3.Discount and profit are negatively correlated.(As Discount increases profit decreases)
4.Discount and sales are negatively correlated.(As Sales increases discount decreases)
5.Nearly no correlation between quantity and discount(zero correlation)

```python
sns.pairplot(data1)
```

From above pairplot, we see that Sales and Profit are positively correlated showing that as Sales increases, Profit also increases. From the Histograms,we learn that the Sales and Quantity variables are heavily right-skewed.

```python
#Pie chart for Category
plt.figure(figsize=(6,6))
textprops={"fontsize":12}
plt.title('Category')
plt.pie(data1['Category'].value_counts(), labels=data1['Category'].value_counts().index,autopct='%1.1f%%',textprops=textprops)
plt.show()
```

From the above pie chart we can see that, We have office supplies category which is having largest demand of 60% and then funrniture and technology.

```python
#Barplot to check profit and sales in each category
data1.groupby('Category')[['Profit','Sales']].agg(sum).plot(kind='bar',figsize=(8,4))
plt.title('Profit and Sales in each Category')
plt.ylabel('Profit And Sales')
plt.xlabel('Category')
plt.show()
```

From above Bar chart we can see that,Category Technology has the highest sales amomg others. And Profit for Techonology and Office Supplies is almost same. And Furniture generates least profit.

```python
#Horizontal bar plot for Sales and profit in each sub-category
data1.groupby('Sub-Category')[['Profit','Sales']].agg(sum).plot(kind='barh',figsize=(12,8))
plt.title('Profit and sales in each sub-category')
plt.xlabel('Profit And Sales')
plt.ylabel('Sub-Category')
plt.show()
```

From above bar plot, We see that, the sub-categories namely Tables, Supplies and Bookcases are in loss. And the sub-categories chairs and phone has done highest sales.

```python
#Bar plot for category wise discount
data1.groupby('Category')['Discount'].agg(sum).plot(kind='bar',figsize=(8,4),color='green')
plt.title('Category wise discount')
plt.ylabel('Discount')
plt.xlabel('Category')
plt.show()
```

From above plot, We see that Discount for Office supplies category was given highest.

```python
#Bar plot for getting state wise sales
data1.groupby('State')['Sales'].agg(sum).plot(kind='bar',figsize=(12,8),color='green')
plt.title('State wise Sales')
plt.ylabel('Sales')
plt.xlabel('State')
plt.show()
```

From above bar plot we see that the State California has the highest sales and also New York.But some states like West virginia,North Dakota ,Maine has done very least sales.

```python
#To check the percent for each segment in each category
plt.figure(figsize=(9,5))
sns.countplot(x='Category', hue='Segment', data=data1)
plt.ylabel('count of segment')
plt.title('Categorywise count of segment')
```

From the above countplot , we see that count of office supply category is higher amongst technology and furniture.
Also, In Office supplies, Segment consumer has done higher purchase. Office supplies category has done highest purchase of each segment amongst other categories Furniture and Technology.
In category, Furniture and Technology purchase count is more or less similar across the segment.

```python
#Count plot for Region wise ship mode
plt.figure(figsize=(9,5))
sns.countplot(x='Region', hue='Ship Mode', data=data1)
plt.ylabel('count of ship mode')
plt.title('Count of Regionwise ship mode')
```

From the above count plot we see that, Least sales are in southern region. Standard class ship mode is more popular in each region amongst other kind of ship mode. 

```python
#Bar plot for getting region wise profit
data1.groupby('Region')['Profit'].agg(sum).plot(kind='bar', color='green')
plt.title('Region wise Profit')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

We can see that Eastern and Westernn region generating more profit. So,Sales should be increase in central and south region.

```python
#Finding the preferable mode of shipping
plt.figure(figsize=(9,5))
sns.countplot(x='Ship Mode', data=data1)
plt.ylabel('count of ship mode')
plt.title('Ship mode preference')
```

Thus, Most preferable mode of shipping is standard class.

```python
plt.figure(figsize=(10,4))
sns.lineplot('Quantity', 'Profit', data=data1, color='y', label='Quantity VS Profit')
plt.legend()
```

From above line plot we see that as Quantity increase profit also increases.

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount', 'Profit', data=data1, color='y', label='Discount VS Profit')
plt.legend()
```

We can  clearly see that from above lineplot that Discount and profit are negatively correlated Means,As Discount increases profit decreases.

```python
data1=data1.groupby('Category').aggregate({
'Sales':'sum',
'Discount':'mean',
'Profit':'sum'    
})
data1
```
